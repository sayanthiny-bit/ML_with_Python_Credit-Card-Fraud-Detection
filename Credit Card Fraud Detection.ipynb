```python
## this data set it very unbalanced so handling unblance 1 label 
##seperate all the narmal transaction and fraudulent transaction 
```

```python
legit = credit_cart_data[credit_cart_data.Class == 0]
fraud = credit_cart_data[credit_cart_data.Class == 1]
```

```python
#handling unblance
#Under-Sampling Method used 
#Build a sample dataset containing similar distriution of normal transaction and Fraudulent Transactions(492)
#492 for normal and fraudlent transaction take for better prediction in ML
```

```python
legit_sample = legit.sample(n=492)
```

```python
# concatenating Two data set legit_sample and fraud
```

```python
new_dataset = pd.concat([legit_sample , fraud], axis = 0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean() ## check whether good or bad sample 
```

```python
## spliting the data into Features $ Targets
```

```python
X = new_dataset.drop(columns = 'Class',axis=1)
Y = new_dataset['Class']
```

```python
print(X)
```

```python
print(Y)
```

```python
# spliting the feture and target data into Training and Test data 
```

```python
X_train, X_test, Y_train, Y_test =train_test_split(X,Y,test_size = 0.2, stratify = Y, random_state =2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

```python
#Model training 
# logistic Regression 
```

```python
model = LogisticRegression(solver='liblinear', max_iter=1000)
```

```python
# training our logisticRegresstion model with training Data 
model.fit(X_train,Y_train)
```

```python
## Model Evaluation based on accuracy score 
```

```python
X_train_prediction = model.predict(X_train) ## find a target value of xtrain data 
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)     ## find a accuracy for ture tatget and find targrt 
```

```python
print('Accuracy on Training Data: ',training_data_accuracy ) 
```

```python
# accuracy of test data 
```

```python
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy on Test Data: ',test_data_accuracy ) 
```

```python
## F1
```

```python
print(classification_report(Y_test, X_test_prediction))
```

#Insights & Analysis
1. High Overall Accuracy:
Model performs well on both training and test sets, indicating it generalizes well and is not overfitting.
Test accuracy of 91.88% is a solid performance, especially for an imbalanced binary classification task.

The F1-score for fraud detection (0.91) shows that the model maintains a strong balance between catching frauds and minimizing false positives.
